Attribute Information from this data set:

1. age
2. sex
3. chest pain type (4 values)
4. resting blood pressure
5. serum cholestoral in mg/dl
6. fasting blood sugar > 120 mg/dl
7. resting electrocardiographic results (values 0,1,2)
8. maximum heart rate achieved
9. exercise induced angina
10. oldpeak = ST depression induced by exercise relative to rest
11. the slope of the peak exercise ST segment
12. number of major vessels (0-3) colored by flourosopy
13. thal: 3 = normal; 6 = fixed defect; 7 = reversable defect
14. the predicted attribute!!

## 5. Modelling 

```python
df.head()
```

```python
# predict target
# split data in features and labels (x and y)

x  = df.drop("target", axis=1)
y = df["target"]
```

```python
# split data into train and test sets
np.random.seed(42)

x_train, x_test, y_train, y_test = train_test_split(x,
                                                   y,
                                                   test_size=0.2)
# len of x_train and y_train is now equal
```

### short Roadmap to build ML Model -> for a classification problem with these three with sklearn

1. logistic regression
2. knearest Neighbours Classifier
3. RandomForest Classifier

for each classifier: 
* train, find patterns on training set
* test it on the test set

```python
# put models in a dict
models = {"Logistic Regression": LogisticRegression(),
         "KNN": KNeighborsClassifier(),
         "Random Forest": RandomForestClassifier()}

# Create a function to fit and score models
def fit_and_score(models, x_train, x_test, y_train, y_test):
    """
    fits and evaluates given ML models
    model: from dict models out of Scikit-Learn models
    x_train: training data (without labels)
    x_test: test data (without labels)
    y_train: training labels
    y_test: test labels
    """
    #set random seed
    np.random.seed(42)
    #create dict for model scores
    model_scores = {}
    #loop through the models
    for name, model in models.items():
        #fit the model to data
        model.fit(x_train, y_train)
        #evaluate the model and append the score to the score dict
        model_scores[name] = model.score(x_test, y_test)
    return model_scores
```

```python
models
```

```python
model_scores = fit_and_score(models, x_train, x_test, y_train, y_test)
model_scores
```

# Hyperparameter Tuning - manual

* research: https://medium.com/@mohtedibf/in-depth-parameter-tuning-for-knn-4c0de485baf6

```python
train_scores = []
test_scores = []

# list of different values of n-neighbours -> look to sklearn knn n_neighbours parameter
neighbors = range(1, 21)
# KNN instance
knn = KNeighborsClassifier()
for i in neighbors:
    knn.set_params(n_neighbors=i)
    knn.fit(x_train, y_train)
    train_scores.append(knn.score(x_train, y_train))
    test_scores.append(knn.score(x_test, y_test))
```

```python
test_scores, train_scores
```

```python
plt.plot(neighbors, train_scores, label="Train Score")
plt.plot(neighbors, test_scores, label="Test Score")
plt.xticks(np.arange(1, 21, 1))
plt.xlabel("Number of neighbors")
plt.ylabel("Model Score")
plt.legend()

print(f"Maximum KNN Score on the test data: {max(test_scores)*100:.2f}%")
```

* here we see n_neighbors has at this time it's best result at 11

## Auto tune hyperparameter and save the best result as new parameter value with RandomizedSearchCV

Research Links:

* https://www.kaggle.com/joparga3/2-tuning-parameters-for-logistic-regression
* https://towardsdatascience.com/logistic-regression-model-tuning-with-scikit-learn-part-1-425142e01af5
* https://machinelearningmastery.com/hyperparameters-for-classification-machine-learning-algorithms/

```python
# create hyperparameter grid for logistics regression
log_reg_grid = {"C": np.logspace(-4, 4, 20),
               "solver": ["liblinear"]}

# create hyperparameter grid for RandomForestClassifier
rf_grid = {"n_estimators": np.arange(10, 1000, 50),
          "max_depth": [None, 3, 5, 10],
          "min_samples_split": np.arange(2, 20, 2),
          "min_samples_leaf": np.arange(1, 20, 2)}
```

```python
# tuning the grids setup with RandomizedSearchCV -> Cross validation (CV)

np.random.seed(42)
rs_log_reg = RandomizedSearchCV(LogisticRegression(),
                               param_distributions=log_reg_grid,
                               cv=5,
                               n_iter=20,
                               verbose=True)

# fit random hyperparameter search model for LogisticRegression
rs_log_reg.fit(x_train, y_train)
```

```python
rs_log_reg.best_params_
```

```python
rs_log_reg.score(x_test, y_test)
```

* LogisticRegression is tuned - now for RandomForestClassifier

```python
np.random.seed(42)

rs_rf = RandomizedSearchCV(RandomForestClassifier(),
                          param_distributions=rf_grid,
                          cv=5,
                          n_iter=20,
                          verbose=True)
rs_rf.fit(x_train, y_train)
```

```python
# find best hyperparameters
rs_rf.best_params_
```

```python
rs_rf.score(x_test, y_test)
```

### calculate evaluation metrics using cross_validation

We're going to calc. precision, recall and f1-score with cross validation (cross_val_score())

```python
cv_acc = np.mean(cv_acc)
cv_acc
```

```python
# cross-validated precision
cv_precision = cross_val_score(clf,
                        x,
                        y,
                        cv=5,
                        scoring="precision")
cv_precision = np.mean(cv_precision)
cv_precision
```

```python
# cross-validated recall
cv_recall = cross_val_score(clf,
                        x,
                        y,
                        cv=5,
                        scoring="recall")
cv_recall = np.mean(cv_recall)
cv_recall
```

```python
# cross-validated f1-score
cv_f1 = cross_val_score(clf,
                        x,
                        y,
                        cv=5,
                        scoring="f1")
cv_f1 = np.mean(cv_f1)
cv_f1
```

```python
# Visualize cross-validated metrics
cv_metrics = pd.DataFrame({"Accuracy": cv_acc,
                          "Precision": cv_precision,
                          "Recall": cv_recall,
                          "F1": cv_f1},
                         index=[0])
cv_metrics.T.plot.bar(title="Cross-validated classification metrics", legend=False)
```
